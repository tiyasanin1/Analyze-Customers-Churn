```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv(r'telco_customer_churn.csv')
df 
```

```python
# using data membeship active
condition=df['tenure']!=0
df=df[condition]
df
```

Data yang digunakan sebanyak 7032 baris

```python
df.info()
```

```python
# focusing on customers who do not have partners or dependents
condition1=df[(df['Partner']=='No') & (df['Dependents']=='No') ]
mean_df=condition1['tenure'].mean()
print(f'Average customers who do not have partners or dependents {mean_df}')
```

```python
#Analyzing customer groups based on contract types
df['TotalCharges']=pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(subset=['TotalCharges'])
result=df.groupby('Contract')['TotalCharges'].agg(['mean','median'])
result
```

```python
# Pivot Table Analysis of Total Charger by Contract and Payment Method
df_pivot_table=df.pivot_table(index='Contract', columns='PaymentMethod', values='TotalCharges', aggfunc='mean')
df_pivot_table
```
